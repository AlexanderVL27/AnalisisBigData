```python
import numpy as np 

for i in range(100):
  
    array = np.random.randint(2, size=(3, 3)) 
    print(array) 

    sumaFila=[sum(i) for i in array]
    print (sumaFila)

    x=0;
    for i in range (len(array)):
        if sumaFila[i]!=0:
            if sumaFila[i]%2 !=0:
                x-=1
            else:
                x+=1      
    if x==len(array):
        print("ES")
    else:
        print("no")
    
```
