```r
#all prime numbers up to given a number

prime_numbers <- function(n)
{
    if(n >= 2)
    {
            x = seq(2,n)
            prime_nums= c()
            for(i in seq(2,n))
            {
                    if(any(x == i)) #Si existe algún número sea igual a i 
                    {
                        prime_nums = c(prime_nums, i)
                        x = c(x[(x%%i)!= 0], i)
                    }#if
            }#for
            return(prime_nums)
    }#if
    else
        stop('Input number should be at least 2.')
}

prime_numbers(15)
```

```r
#Find the factors of a given number

print_factors = function(n)
{
    print(paste('The factors of',n,'are:'))
    for(i in 1:n)
    {
        if((n%%i)==0)
        {
                print(i)
        }#if
    }#for
}#function

print_factors(4)
```
