**La primeragráfica es una grafica de pastel la cual representa relación de películas que están o no en una colección**

```python
#Graficas
y = 0;
x = 0;
for i in datos.belongs_to_collection:
    if pd.isna(i)==True:
        x = x + 1
    else:
        y = y + 1
        
posibilidades = ['En coleccion','No coleccion']
slices = [y,x]
```

```python
#colores
color = ['r', 'y']

plt.pie(slices, labels = posibilidades , colors=color,startangle = 90, shadow = True,explode = (0,0.1),radius = 1.2,autopct = '%1.1f%%')

plt.legend()

plt.show()
```

**Grafica de barras**

```python
names = np.array(datos.title)
nam = np.array(names[0:15])

y_pos = np.arange(len(nam))
arrayofdata = np.array(datos.vote_average)
arrayofdata2 = np.array(arrayofdata[0:15])

plt.barh(y_pos,arrayofdata2,align = 'center')
plt.yticks(y_pos,nam)
plt.xlabel('Promedio')
plt.title('Votacion Promedio')

plt.show()
```
