```r
limpiar_tokenizar <- function(texto){
nuevo_texto <- tolower(texto)
nuevo_texto <- str_replace_all(nuevo_texto,"http\\S*", "") # 2
nuevo_texto <- str_replace_all(nuevo_texto,"[[:punct:]]", " ") # 3
nuevo_texto <- str_replace_all(nuevo_texto,"[[:digit:]]", " ") # 4
nuevo_texto <- str_replace_all(nuevo_texto,"[\\s]+", " ")  # 5 
nuevo_texto <- str_replace_all(nuevo_texto,"[[:emoji:]]", " ")
nuevo_texto <- str_split(nuevo_texto, " ")[[1]]   #6
nuevo_texto <- keep(.x = nuevo_texto, .p = function(x){str_length(x) > 1})
    return(nuevo_texto)
}
# 1) Se aplica la funci√≥n de limpieza y 
# tokenizaci√≥n a cada tweet
tweets1 <- tweets1 %>% mutate(texto_tokenizado = map(.x = texto, .f = limpiar_tokenizar))
#tweets1 %>% select(texto_tokenizado) %>% head()

#write.csv(tweets1,"datos.csv")
# 1) Al tokenizar, el elemento de estudio ha pasado a 
# ser cada token, incumpliendo as√≠ la condici√≥n de¬†tidy data. 

tweets_tidy <- tweets1 %>% select(-texto) %>% unnest()
tweets_tidy <- tweets_tidy %>% rename(token = texto_tokenizado)
tweets_tidy
```

```r

# Practica 2 - Clasificaci√≥n de tweets:

# Para poder aplicar algoritmos de clasificaci√≥n 
# a un texto, es necesario crear una representaci√≥n 
# num√©rica del mismo
tweets_auron_mariana <- tweets1 %>% filter(autor %in% c("auronplay", "elmarianaa"))
set.seed(123)
train <- sample(x = 1:nrow(tweets_auron_mariana), size= 0.5*nrow(tweets_auron_mariana))
tweets_train <- tweets_auron_mariana[train, ]
tweets_test  <- tweets_auron_mariana[-train, ]

# Es importante verificar que la proporci√≥n de 
# cada grupo es similar en el set de entrenamiento 
# y en el de test.
table(tweets_train$autor) / length(tweets_train$autor)
table(tweets_test$autor) / length(tweets_test$autor)

# Limpieza y tokenizaci√≥n de los documentos de entrenamiento
tweets_train$texto <- tweets_train$texto %>% map(.f = limpiar_tokenizar) %>% map(.f = paste, collapse = " ") %>% 
unlist()

tweets_train

```

```r
# Creaci√≥n de la matriz documento-t√©rmino
matriz_tfidf_train <- dfm(x = tweets_train $ texto, remove = lista_stopwords)
matriz_tfidf_train

# Se reduce la dimensi√≥n de la matriz eliminando aquellos 
# t√©rminos que aparecen en menos de 5 documentos. Con esto se consigue eliminar ruido.
matriz_tfidf_train <- dfm_trim(x = matriz_tfidf_train, min_docfreq = 5)

# Conversi√≥n de los valores de la matriz a tf-idf
matriz_tfidf_train <- dfm_tfidf(matriz_tfidf_train, scheme_tf = "prop",scheme_df = "inverse")
matriz_tfidf_train

# Limpieza y tokenizaci√≥n de los documentos de test
tweets_test$texto <- tweets_test$texto %>% map(.f = limpiar_tokenizar) %>%map(.f = paste, collapse = " ") %>% unlist()
# Identificaci√≥n de las dimensiones de la matriz de entrenamiento
# Los objetos dm() son de clase S4, se accede a sus elementos mediante @
dimensiones_matriz_train <- matriz_tfidf_train@Dimnames$features
# Conversi√≥n de vector a diccionario pasando por lista
dimensiones_matriz_train <- as.list(dimensiones_matriz_train)
names(dimensiones_matriz_train) <- unlist(dimensiones_matriz_train)
dimensiones_matriz_train <- dictionary(dimensiones_matriz_train)

# Proyecci√≥n de los documentos de test
matriz_tfidf_test<-dfm(x=tweets_test$texto,dictionary=dimensiones_matriz_train)
matriz_tfidf_test<-dfm_tfidf(matriz_tfidf_test, scheme_tf="prop",scheme_df = "inverse")
matriz_tfidf_test


```

```r
# Como modelo de predicci√≥n se emplea una M√°quina de soporte vectorial¬†(SVM). 
# Este m√©todo de aprendizaje estad√≠stico suele dar buenos resultados en clasificaci√≥n.
# Ajuste del modelo
modelo_svm <- svm(x = matriz_tfidf_train, y = as.factor(tweets_train$autor),kernel = "linear", cost = 1, scale = TRUE, type = "C-classification")
modelo_svm
```

```r
# Predicciones
predicciones <- predict(object = modelo_svm, newdata = matriz_tfidf_test)
```

```r

# Error de predicci√≥n
# Matriz de confusi√≥n
table(observado = tweets_test$autor, predicho = predicciones)

# Error de clasificaci√≥n
clasificaciones_erroneas <- sum(tweets_test$autor != predicciones)
error <- 100 * mean(tweets_test$autor != predicciones)
paste("N√∫mero de clasificaciones incorrectas =", clasificaciones_erroneas)
paste("Porcentaje de error =", round(error,2), "%")
```

```r
# Optimizaci√≥n de hiperpar√°metros
# El m√©todo de SVM lineal tiene un √∫nico hiperpar√°metro C 
# que establece la penalizaci√≥n por clasificaci√≥n incorrecta, 
# regulando as√≠ el balance entre bias y varianza. 
# Al tratarse de un hiperpar√°metro, su valor √≥ptimo 
# no se aprende en el proceso de entrenamiento, 
# para estimarlo hay que recurrir a validaci√≥n cruzada.
set.seed(369)
svm_cv <- tune("svm", train.x =  matriz_tfidf_train,
               train.y = as.factor(tweets_train$autor),
               kernel = "linear", ranges = list(cost = c(0.1, 0.5, 1, 2.5, 5)))
summary(svm_cv)
ggplot(data = svm_cv$performances, aes(x = cost, y = error)) +
  geom_line() +
  geom_point() +
  geom_errorbar(aes(ymin = error - dispersion, ymax = error + dispersion)) +
  theme_bw()
  
# Acorde al error estimado por validaci√≥n cruzada, 
# el valor √≥ptimo del hiperpar√°metro  C es 0.5. 
# Se reajusta el modelo con este valor.

svm_cv <- svm(x = matriz_tfidf_train, y = as.factor(tweets_train$autor),
                  kernel = "linear", cost = 0.5, scale = TRUE)

predicciones <- predict(object = svm_cv, newdata = matriz_tfidf_test)
table(observado = tweets_test$autor, predicho = predicciones)

clasificaciones_erroneas <- sum(tweets_test$autor != predicciones)
error <- 100 * mean(tweets_test$autor != predicciones)
paste("N√∫mero de clasificaciones incorrectas =", clasificaciones_erroneas)

paste("Porcentaje de error =", round(error,2), "%")

# Empleando un modelo de SVM lineal con hiperpar√°metro C=0.5 
# se consigue un test error del 3.51%. Se trata de un 
# porcentaje de error bastante bajo, aun as√≠, explorando 
# otros modelos como pueden ser SVM no lineales o Random 
# Forest podr√≠an alcanzarse mejores resultados. 
```

```r
# ---------------------------------------------------
# Practica 3 - An√°lisis de sentimientos: 

# Una forma de analizar el sentimiento de un de un texto
# es considerando su sentimiento como el la suma de los 
# sentimientos de cada una de las palabras que lo forman. 
# Esta forma consigue un buen equilibro entre complejidad 
# y resultados.

# Para llevar a cabo esta aproximaci√≥n es necesario disponer 
# de un diccionario en el que se asocie a cada palabra un 
# sentimiento o nivel de sentimiento. A estos diccionarios 
# tambi√©n se les conoce como sentiment lexicon. 
# El paquete tidytext contiene 3 diccionarios distintos:

# AFINN: asigna a cada palabra un valor entre -5 y 5. 
# Siendo -5 el m√°ximo de negatividad y +5 el m√°ximo de 
# positividad.

# bing: clasifica las palabras de forma binaria 
# positivo/negativo.

# nrc: clasifica cada palabra en uno o m√°s de los 
# siguientes sentimientos: positive, negative, 
# anger, anticipation, disgust, fear, joy, sadness, 
# surprise, and trust.

# ******************************************
# fuente: https://rpubs.com/jboscomendoza/analisis_sentimientos_lexico_afinn
download.file("https://raw.githubusercontent.com/jboscomendoza/rpubs/master/sentimientos_afinn/lexico_afinn.en.es.csv","lexico_afinn.en.es.csv")
sentimientos <-read.csv("lexico_afinn.en.es.csv", stringsAsFactors = F,fileEncoding = "latin1") %>% tbl_df()
sentimientos
# ******************************************
```

```r
# Veamos nuestros resultados con ggplot()

tweets_sent_fecha_2 %>%
  ggplot() +
  aes(Fecha, Media, color = autor) +
  geom_line() +
  theme(legend.position = "top")

# No nos dice mucho. Sin embargo, 
# si separamos las l√≠neas por candidato, 
# usando facet_wrap(), ser√° m√°s f√°cil observar 
# el las tendencias de los Candidatos.

tweets_sent_fecha_2 %>%
  ggplot() +
  aes(Fecha, Media, color = autor) +
  geom_hline(yintercept = 0, alpha = .35) +
  geom_line() +
  facet_grid(autor~.) +
  theme(legend.position = "none")
```
