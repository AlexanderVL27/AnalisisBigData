```r
#Expresiones condicionales, repeticiones y lógica
x <- 3
y <- NULL
if (x > 5) { y <- x * 3 } else { y <- 6 - x }
y

z <- NULL
if (x %% 2 == 1 & y %% 2 == 1) { z <- 2 } else { z <- 3 }
z

if (z != 1) { print("no es uno") }

if (x == 2 | y == 2 | z == 2) { print("hay un dos") } else { print("no hay un dos") }

for (i in 1:5) { print(2**i) }

while (x > 0) { print("quito uno"); x <- x - 1 }
x
```

```r
#Subrutinas
potdos <- function(x) 2**x
potdos(3)

eucl <- function(x1, y1, x2, y2) { dx <- x1 - x2; dy <- y1 - y2; return(sqrt(dx^2 + dy^2)) }
eucl(1, 2, 3, 4)
```

```r
#Ejemplo base de datos con csv
datos <- read.csv("../input/conjunto-de-datos/ejemplo.csv", header=TRUE, sep=" ", stringsAsFactors=FALSE)
datos
datos$Estatura <- as.numeric(datos$Estatura) #as.numeric para convertir las strings a numeric
datos$Gatos <- as.numeric(datos$Gatos)
datos$Perros <- as.numeric(datos$Perros)
sum(datos$Gatos) #Sumar la columna; el total de gatos
sum(datos$Gatos) > sum(datos$Perros)
datos$Gatos == datos$Perros

datos[datos$Nombre == "Elisa",] #Mostrar los registos donde coincida el nombre
prom <- mean(datos$Estatura)
datos[datos$Estatura > prom,]

datos <- rbind(datos, c("Tania", 175, 0, 2)) #rbind se usa para combinar vectores 'agregar un registro'
datos
#Agrega una columna nueva
datos$Estudiante <- c(FALSE, TRUE, TRUE, TRUE, FALSE, FALSE) #settear los datos de Estudiante
datos
#Selecciona los registros por rangos
datos[3,]
datos[2:4,]
datos[2:4,3]
datos$Estudiante <- as.factor(datos$Estudiante) #Te da los niveles es decir las opciones.
datos$Estudiante
datos$Nombre <- as.factor(datos$Nombre)
datos$Nombre
```
